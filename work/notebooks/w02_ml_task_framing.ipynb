## 5. Why ML beats a fixed rule here

Machine learning can consider multiple observable signals together, including impressions, clicks, average position, engagement, content freshness, and other metrics. Because these signals interact in complex ways, a ranking model can provide better decision support than relying on a single manually defined rule.
